```python
import pandas as pd
import statsmodels.api as sm
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/baren-kumar/Data-Sheet/refs/heads/main/sheet_metal_production_data.csv')
data.head()
```

```python
# Define independent variables (processes) and dependent variable (Total Cost)
X = data[['Shearing', 'Blanking', 'Punching', 'Bending', 'Deep_Drawing', 'Forming',
          'Trimming', 'Deburring', 'Welding', 'Surface_Cleaning', 'Painting',
          'Quality_Control', 'Packaging']]
y = data['Total_Cost']

# Add a constant for the intercept
X = sm.add_constant(X)

# Fit the multiple linear regression model
model = sm.OLS(y, X).fit()

# Print the full model summary
print(model.summary())

# Extract and print key results in a table format
coefficients = model.params
p_values = model.pvalues
conf_int = model.conf_int()

results_df = pd.DataFrame({
    'Coefficient': coefficients,
    'P-value': p_values,
    '95% CI Lower': conf_int[0],
    '95% CI Upper': conf_int[1]
})

print("\nRegression Coefficients Table:")
print(results_df)

# Interpretation example
print("\nInsights:")
print("High-impact processes (largest coefficients): Painting, Deep Drawing, Bending.")
print("Focus on these for cost reduction.")
```
